```python
source_path = r'C:\xampp\htdocs\PA\PA_KB_B2_20_5\raw_datasets'
destination_path = r'C:\xampp\htdocs\PA\PA_KB_B2_20_5\datasets'
test_percentage = 0.15
```

```python
import os
import json
import random
from pathlib import Path
```

```python
datasets_walker = os.walk(source_path)

labels_list = next(datasets_walker)[1]
files_list = dict()

for i in range(len(labels_list)):
    _, __, files_ = next(datasets_walker)
    random.shuffle(files_)
    files_list[labels_list[i]] = files_

print(json.dumps(files_list, indent=4))
print(len(files_list.keys()))
```

```python
datasets_splits = dict()

for i, key in enumerate(files_list.keys()):
    files_count = len(files_list.get(key))
    os.makedirs(os.path.join(destination_path, fr'train\{key}'))
    os.makedirs(os.path.join(destination_path, fr'val\{key}'))
    # training data
    for file in files_list.get(key)[int(files_count*test_percentage):]:
        Path(f'{source_path}/{key}/{file}').rename(fr'{destination_path}\train\{key}\{file}')
    # test data
    for file in files_list.get(key)[:int(files_count*test_percentage)]:
        Path(f'{source_path}/{key}/{file}').rename(fr'{destination_path}\val\{key}\{file}')
```
