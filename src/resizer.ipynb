```python
import matplotlib.pyplot as plt
import os
from PIL import Image, ImageOps
```

### Demonstration

```python
# Open images on file
img = Image.open(r'C:\xampp\htdocs\PA\PA_KB_B2_20_5\raw_datasets\Sedan\01584_Rolls-Royce Ghost Sedan 2012.jpg')
plt.figure(1, (5,5))
plt.xlabel('Original Image')
plt.imshow(img)
# get image size and calculate shrink percentage
img_width, img_height = img.size
shrink_by = 128 / img_width
print(f'{shrink_by=}')
print(f'original img size = {img.size}')
# turn image into grayscale
img = ImageOps.grayscale(img)
# resize image respecting it's original aspect ratio
img.thumbnail((int(img_width * shrink_by), int(img_height * shrink_by)), Image.Resampling.LANCZOS)
plt.figure(2, (5,5))
plt.imshow(img, cmap='gray')
plt.xlabel('Processed Image')
print(f'resized img size = {img.size}')
```

### Resize and Grayscale Image

```python
# file walker
datasets_walker = os.walk(datasets_path)

labels_list = next(datasets_walker)[1]

# get every class folder
for i in range(len(labels_list)):
    print(f'{i}: {labels_list[i]}')
    # get next walker
    dir_, __, files_ = next(datasets_walker)
    for file in files_:
        # Open images on file
        img = Image.open(os.path.join(dir_, file))
        # get image size and calculate shrink percentage
        img_width, img_height = img.size
        shrink_by = 128 / img_width
        # turn image into grayscale
        img = ImageOps.grayscale(img)
        # resize image respecting it's original aspect ratio
        img.thumbnail((int(img_width * shrink_by), int(img_height * shrink_by)), Image.Resampling.LANCZOS)
        # save the result
        img.save(os.path.join(dir_, file))
```
