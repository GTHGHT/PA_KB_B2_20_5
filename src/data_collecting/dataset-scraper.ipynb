# Dataset Scraper / Crawler

I put this notebook on a datasets (in this case it's [this datasets](https://www.kaggle.com/datasets/ssturua/myautoge-cars-dataset)) then it would gather up all the image into the output directory. I archive the output into a zip where i can download it to my device

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

ds = pd.read_csv('../input/myautoge-cars-dataset/characteristics.csv')
ds.head()
```

```python
#     &(ds['model']=='Combo')&
#     (ds['manufacturer']=='OPEL')
ds[ds['category'] == 'Sedan']
```

```python
#     &(ds['model']=='Doblo')&
#     (ds['manufacturer']=='FIAT')
car_id = ds[ds['category'] == 'Sedan']['ID']
car_id.head()
```

```python
import shutil
shutil.rmtree("/kaggle/working")
```

```python
from PIL import Image
import os
image_path = '../input/myautoge-cars-dataset/images/images'
for i, car in enumerate(car_id):
    if i >= 203:
        break
    os.makedirs(f'./{car}')
    try:
        print(os.path.join(image_path, str(car)))
        for file in os.listdir(os.path.join(image_path, str(car))):
            img = Image.open(os.path.join(image_path, f"{car}/{file}"))
            img.save(os.path.join(f'./{car}', file))
    except Exception as e:
        print(e)
```
